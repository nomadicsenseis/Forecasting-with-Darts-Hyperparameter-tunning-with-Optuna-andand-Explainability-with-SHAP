```python
import boto3
import pandas as pd

def read_csv_files_from_s3(s3_path):
    # Initialize Boto3 client for S3
    s3_client = boto3.client('s3')

    # Extract bucket name and prefix from the S3 path
    bucket, prefix = s3_path.replace('s3://', '').split('/', 1)

    # Get a list of all keys (object paths) under the specified S3 path
    s3_resource = boto3.resource('s3')
    s3_keys = [item.key for item in s3_resource.Bucket(bucket).objects.filter(Prefix=prefix)]

    # Generate a list of full S3 paths for all CSV files
    preprocess_paths = [f"s3://{bucket}/{key}" for key in s3_keys]

    # Read all CSV files into a list of DataFrames
    dfs = [pd.read_csv(s3_client.get_object(Bucket=bucket, Key=key)['Body']) for key in s3_keys]

    # Concatenate DataFrames into a single DataFrame
    df_concatenated = pd.concat(dfs, axis=0, ignore_index=True)

    return df_concatenated
```

## Data summary

```python
def get_summary(df):
    
    """Data summary returns a table with the following information: Variable, Data type, Null count, Null count %, Unique count, Unique values"""

    summary_df = pd.DataFrame()
    
    n_rows = df.shape[0]
    
    column_list = df.columns
    for col in column_list:
        data_type = df[col].dtype
        null_count = df[col].isna().sum()
        unique_values = np.sort(df.loc[df[col].notnull(), col].unique().tolist())
        unique_count = df[col].nunique() 
        
        col_info = pd.DataFrame(data = {"Variable": [col], 
                                        "Data type": [data_type], 
                                        "Null count": [null_count], 
                                        "Null count %": [np.round(null_count/n_rows*100, 1)],
                                        "Unique count": [unique_count],
                                        "Unique values": [unique_values]})
        
        summary_df = pd.concat([summary_df, col_info], axis = 0)
            
    first_date = df["date_flight_local"].min()
    last_date = df["date_flight_local"].max()
    print("N rows: ", n_rows)
    print("Flight dates: ", first_date, " - ", last_date)
    print("Columns: ", df.columns)
    return summary_df.reset_index(drop = True)
```

```python
def create_flight_time_features(df):
    
    #### REVIEWD
    # Create new columns of departure datetime and arrival datetime to use from the related columns of surveys_maritz
    df["actual_dep_datetime"] = df["time_real_dep_gmt"]
    df["scheduled_dep_datetime"] = df["time_sched_dep_gmt"]
    
    df["actual_arr_datetime"] = df["time_real_arr_gmt"]
    df["scheduled_arr_datetime"] = df["time_sched_arr_gmt"]
    
    # Impute missing values with the information from the table flight
    row_filter = df.loc[(df["date_flight_local"] < '2019-08-27')  # Si la fecha es anterior al 27/08/2019 y cualquier de los 4 campos está vacío en surveys_maritz...
                       | (df["time_real_dep_gmt"].isna()) 
                       | (df["time_sched_dep_gmt"].isna())
                       | (df["time_real_arr_gmt"].isna()) 
                       | (df["time_sched_arr_gmt"].isna())].index
    
    # ... rellenar/sobreescribir los 4 campos con la información de la tabla flight
    df.iloc[row_filter]["actual_dep_datetime"] = df.iloc[row_filter]["utc_dep_time_actual"]
    df.iloc[row_filter]["scheduled_dep_datetime"] = df.iloc[row_filter]["utc_dep_time_scheduled"]
    
    df.iloc[row_filter]["actual_arr_datetime"] = df.iloc[row_filter]["utc_arr_time_actual"]
    df.iloc[row_filter]["scheduled_arr_datetime"] = df.iloc[row_filter]["utc_arr_time_scheduled"]
    
    # Calculate delay in depature and arrival in minutes
    df["dep_delay_min"] = (df["actual_dep_datetime"] - df["scheduled_dep_datetime"])/timedelta(minutes=1)
    df["arr_delay_min"] = (df["actual_arr_datetime"] - df["scheduled_arr_datetime"])/timedelta(minutes=1)
    
    # Calculate flight duration
    df["flight_duration_min"] = (df["scheduled_arr_datetime"] - df["scheduled_dep_datetime"])/timedelta(minutes=1)
    
    # Extract hour and minute from departure datetime
    df["local_dep_datetime"] = df["time_sched_dep_local"]
    row_filter = df.loc[(df["local_dep_datetime"].isna())].index
    
    df.iloc[row_filter]["local_dep_datetime"] = df.iloc[row_filter]["loc_dep_time_scheduled"]
    
    df["actual_dep_hour_min"] = df["local_dep_datetime"].apply(lambda x: x if pd.isna(x) else timedelta(hours = x.hour, minutes = x.minute))

    # Group departure hour and minute in hour ranges
    df.loc[(df["actual_dep_hour_min"] >= timedelta(hours = 7, minutes = 0)) 
                       & (df["actual_dep_hour_min"] <= timedelta(hours = 11, minutes = 59)), "departure_time_range"] = "From 7:00 to 11:59"
    
    df.loc[(df["actual_dep_hour_min"] >= timedelta(hours = 12, minutes = 0)) 
                       & (df["actual_dep_hour_min"] <= timedelta(hours = 17, minutes = 59)), "departure_time_range"] = "From 12:00 to 17:59"
    
    df.loc[(df["actual_dep_hour_min"] >= timedelta(hours = 18, minutes = 0)) 
                       & (df["actual_dep_hour_min"] <= timedelta(hours = 22, minutes = 59)), "departure_time_range"] = "From 18:00 to 22:59"
    
    df.loc[(df["actual_dep_hour_min"] >= timedelta(hours = 23, minutes = 0)) 
                       | (df["actual_dep_hour_min"] <= timedelta(hours = 6, minutes = 59)), "departure_time_range"] = "From 23:00 to 6:59"
```

```python
def group_residence_country(df):
    
    ESP = ["ES"]
    
    EUR = ["AT", "BE", "CH", "DE", "FR", "GB", "UX", "GR", "IL", "IT", "MA", "NL", "PT", "RU", "SE", "CZ", "DZ", 
                 "IE", "SI", "AD", "DK", "FI", "HR", "HU", "LU", "PL", "RO", "SK", "MC", "EE", "NO", "SN"]    
    
    AM_NOR = ["CA", "MX", "PR", "US"]

    LATAM = ["CR", "CU", "DO", "GT", "NI", "PA", "SV", "VE", "HN", "AR", "BR", "CL", "CO", "EC", "PE", "UY", "BO"]
    
    df.loc[df["res100_country_code_survey"].isin(ESP), "country_aggr"] = "ESP"
    df.loc[df["res100_country_code_survey"].isin(EUR), "country_aggr"] = "EUR"
    df.loc[df["res100_country_code_survey"].isin(AM_NOR), "country_aggr"] = "AM_NOR"
    df.loc[df["res100_country_code_survey"].isin(LATAM), "country_aggr"] = "LATAM"
    
    df["country_aggr"] = df["country_aggr"].fillna("Others")
```

```python
def group_fleet(df):
    
    # Create the aggregated fleet column copying 'fleet'
    df["fleet_aggr"] = df["fleet"] 
    
    # Assign group using the following criteria:
    df.loc[df["operating_airline_code"].isin(["YW"]), "fleet_aggr"] = "YW"
    df.loc[df["fleet_aggr"].isin(["319","31A","31B"]), "fleet_aggr"] = "319"
    df.loc[df["fleet_aggr"].isin(["320","32A","32S"]), "fleet_aggr"] = "320"
    df.loc[df["fleet_aggr"].isin(["321","32B","32D"]), "fleet_aggr"] = "321"
    df.loc[df["fleet_aggr"].isin(["345","346"]), "fleet_aggr"] = "346"
    ## Other
    df.loc[df["fleet_aggr"].isin(["32W","739","75F","789","ATX","CKH","CRK","CX"]), "fleet_aggr"] = "Other"
    df.loc[df["fleet_aggr"].isna(), "fleet_aggr"] = "Other"
```

```python
def issues_binary(df, dict_touchpoints):
    all_unique_issues = [i for i in dict_touchpoints["issue_type_2"].unique()]
    for issue in all_unique_issues:
        df[issue] = df[issue].apply(lambda x: 0 if pd.isna(x) else 1)

    for touchpoint in dict_touchpoints["associated_touchpoint"].unique():
        related_issues = [i for i in dict_touchpoints.loc[dict_touchpoints["associated_touchpoint"] == touchpoint]["issue_type_2"]]
        new_feature = "issues_{}".format(touchpoint)
        df[new_feature] = df[related_issues].sum(axis = 1)
        df[new_feature] = df[new_feature].apply(lambda x: 1 if x >= 1 else 0)
```

```python
def input_df_preparation(input_df, columns_selected):
    
  # Prepare numeric columns  
    ## Fill null values with np.nan
    numeric_columns = input_df.select_dtypes(include = np.number).columns # Recognize numeric columns 
    input_df[numeric_columns] =  input_df[numeric_columns].fillna(np.nan)
    
    ##################################################################
    # Satisfaction
    ##################################################################
    imp_mean = SimpleImputer(missing_values=np.nan, strategy='median')
    
    default_questions = ['bkg_100_booking', 'bkg_200_journey_preparation', 'pfl_100_checkin', 'pfl_200_security', 'pfl_500_boarding', 
                        'ifl_300_cabin', 'ifl_200_flight_crew_annoucements', 'ifl_400_food_drink', 'ifl_100_cabin_crew', 'arr_100_arrivals', 
                        'pun_100_punctuality', 'img_310_ease_contact_phone', 'img_320_ease_contact_ibplus_mail']
    input_df[default_questions] = imp_mean.fit(input_df[default_questions]).transform(input_df[default_questions])
    
    
    row_filter = input_df.loc[input_df["inm_050_issues_t_f"] == "Yes"].index.tolist()
    feature_to_impute = "inm_400_issues_response" 
    if len(row_filter) > 0:
        input_df.iloc[row_filter][[feature_to_impute]] = imp_mean.fit(input_df[[feature_to_impute]]).transform(input_df.iloc[row_filter][[feature_to_impute]])
    

    row_filter = input_df.loc[input_df["lvl_loyalty_program_survey"] != "Not a member"].index.tolist() 
    feature_to_impute = "loy_200_loyalty_programme" 
    if len(row_filter) > 0:    
        input_df.iloc[row_filter][[feature_to_impute]] = imp_mean.fit(input_df[[feature_to_impute]]).transform(input_df.iloc[row_filter][[feature_to_impute]])
    
    
    row_filter = input_df.loc[input_df["date_flight_local"] >= "2020-07-11"].index.tolist()
    feature_to_impute = "cov_300_appropiate_changes_to_reassure"
    if len(row_filter) > 0:    
        input_df.iloc[row_filter][[feature_to_impute]] = imp_mean.fit(input_df[[feature_to_impute]]).transform(input_df.iloc[row_filter][[feature_to_impute]])
    
    
    row_filter = input_df.loc[input_df["cla_400_lounge_t_f"] == "Yes"].index.tolist()
    feature_to_impute = "pfl_300_lounge"
    if len(row_filter) > 0:    
        input_df.iloc[row_filter][[feature_to_impute]] = imp_mean.fit(input_df[[feature_to_impute]]).transform(input_df.iloc[row_filter][[feature_to_impute]])
    
    
    row_filter = input_df.loc[(input_df["cla_300_connection_from"].notnull()) | (input_df["cla_300_connection_to"].notnull())].index.tolist()
    feature_to_impute = "con_100_connections"
    if len(row_filter) > 0:
        input_df.iloc[row_filter][[feature_to_impute]] = imp_mean.fit(input_df[[feature_to_impute]]).transform(input_df.iloc[row_filter][[feature_to_impute]])
    
    
    row_filter = input_df.loc[(input_df["cla_500_ife_t_f"].notnull()) & (input_df["cla_500_ife_t_f"] != "No - I did not want to use it")].index.tolist() 
    feature_to_impute = "ifl_500_ife"
    if len(row_filter) > 0:    
        input_df.iloc[row_filter][[feature_to_impute]] = imp_mean.fit(input_df[[feature_to_impute]]).transform(input_df.iloc[row_filter][[feature_to_impute]])
    
    
    row_filter = input_df.loc[((input_df["date_flight_local"] >= '2019-07-08') & (input_df["cla_610_wifi_aware"].notnull()) & (input_df["cla_610_wifi_aware"] != "No"))
                             | ((input_df["date_flight_local"] < '2019-07-08') & (input_df["cla_610_wifi_aware"].notnull()))].index.tolist()  
    feature_to_impute = "ifl_600_wifi"
    if len(row_filter) > 0:    
        input_df.iloc[row_filter][[feature_to_impute]] = imp_mean.fit(input_df[[feature_to_impute]]).transform(input_df.iloc[row_filter][[feature_to_impute]])
    
                                
    row_filter = input_df.loc[(input_df["dig_100_web_mobile_t_f"] == "Yes - mobile app") | (input_df["dig_100_web_mobile_t_f"] == "Yes - Website and mobile app")].index.tolist()
    feature_to_impute = "dig_300_web"
    if len(row_filter) > 0:    
        input_df.iloc[row_filter][[feature_to_impute]] = imp_mean.fit(input_df[[feature_to_impute]]).transform(input_df.iloc[row_filter][[feature_to_impute]])
    
#### Covid / No Covid
    input_df.loc[input_df["date_flight_local"] <= "2020-03-13", "indicator_covid"] = 0
    input_df.loc[input_df["date_flight_local"] > "2020-03-13", "indicator_covid"] = 1
    
### Impute flight duration and delay in minutes
    #feature_to_impute = ["flight_duration_min", "dep_delay_min", "arr_delay_min", "load_factor"]
    #input_df[feature_to_impute] = imp_mean.fit(input_df[feature_to_impute]).transform(input_df[feature_to_impute])
    
    ### Select columns
    input_df = input_df[columns_selected]
    #numeric_columns = input_df.select_dtypes(include = np.number).columns # Recognize numeric columns 
    categorical_columns = input_df.select_dtypes(include = object).columns # Recognize catgeorical columns    
    
    # Prepare categorical columns
    if len(categorical_columns) >= 1:
    
        # For each categorical column, generate n - 1 binary columns, being n the number of unique values of that column
        for var in categorical_columns:

            df_dummies = pd.get_dummies(input_df[var].fillna("Not informed"))
            new_colnames = {}
            
            for i in df_dummies.columns:
                new_colnames[i] = var + "_" + i
            df_dummies = df_dummies.rename(columns = new_colnames)
            input_df = pd.concat((input_df.drop([var], axis = 1), df_dummies[df_dummies.columns[:-1]]), axis = 1)  
            
    return input_df
```

### Train, test, and predict

```python
def train_test_predict(model, df, input_vars, target_vars, train_start_end_date, test_start_end_date):
    
    print("Targets : {}".format(target_vars))
    
    result_dict = {}
    
    # Select columns and prepara data
    columns_selected = input_vars + target_vars
    
    df_processed = input_df_preparation(df, columns_selected = columns_selected + ["date_flight_local", "respondent_id"])
    
    X_train, X_test, y_train, y_test = train_test_split(df_processed.drop(target_vars, axis = 1), df_processed[target_vars],
                                                        stratify=df_processed["date_flight_local"].dt.year, 
                                                        test_size=0.2,
                                                        random_state = 1)
    
    train_set_to_save = pd.concat([X_train, y_train], axis = 1)
    test_set_to_save = pd.concat([X_test, y_test], axis = 1)
    
    train_to_drop = X_train.loc[X_train["date_flight_local"].dt.year == 2022].index
    test_to_drop = X_test.loc[X_test["date_flight_local"].dt.year == 2022].index
    
    X_train = X_train.drop(train_to_drop)
    y_train = y_train.drop(train_to_drop)
    X_test = X_test.drop(test_to_drop)
    y_test = y_test.drop(test_to_drop)    
    
    for target in target_vars:
    
        print("Training model for: {}".format(target))
        
        X_train_model = X_train.drop(["date_flight_local", "respondent_id"], axis = 1)
        X_test_model = X_test.drop(["date_flight_local", "respondent_id"], axis = 1)
        
        y_train_model = y_train[target]
        y_test_model = y_test[target]
        
        clf = model.copy()
        
        clf= clf.fit(X_train_model, y_train_model)
        train_pred = clf.predict(X_train_model)
        train_pred_prob = clf.predict_proba(X_train_model) 
        
        test_pred = clf.predict(X_test_model)
        test_pred_prob = clf.predict_proba(X_test_model) 
  
        ## Store the results in the dictionatiory
        result_dict[target] = {"train": {"model": clf, "y_true": y_train_model, "y_pred": train_pred, "y_prob": train_pred_prob}, 
                              "test": {"model": clf, "y_true": y_test_model, "y_pred": test_pred, "y_prob": test_pred_prob}} 
    
    ## Store the datasets as well
    result_dict["train_set"] = train_set_to_save
    result_dict["test_set"] = test_set_to_save
    #result_dict["train_set_index"] = train_set_index_to_save
    #result_dict["test_set_index"] = test_set_index_to_save
    result_dict["train_start_end_date"] = train_start_end_date
    result_dict["test_start_end_date"] = test_start_end_date
        
    return result_dict, clf
```

### Calculate classification scores

```python
def calculate_classication_scores(y_true, y_pred, y_prob):

    """ Use classfication measures to evaluate the model."""
    
    sample_size = y_true.count()
    pct_target = y_true.sum()/y_true.count()
    pct_pred = np.sum(y_pred)/len(y_pred)
    cm = confusion_matrix(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_prob[:, 1])
    f1 = f1_score(y_true, y_pred)
    loss = log_loss(y_true, y_prob)
    
    score_dict = {"sample size": sample_size,
                  "pct_target": [pct_target], "pct_pred": [pct_pred],
                  "confusion_matrix": [cm], 
                  "accuracy": [accuracy], "precision": [precision], "recall": [recall], "f1": [f1],
                  "roc_auc": [roc_auc],
                  "log_loss": [loss]}
    
    score_df = pd.DataFrame.from_dict(score_dict)
    return score_df
```

```python
def display_calculation_score(result_dict, model_name, dataset_name):
    
    """ This functions calls 'calculate_classication_scores' function and returns the scores for the given models in a dataframe."""
    
    result_df = pd.DataFrame()
    for target in ["promoter_binary", "detractor_binary"]:
        for dataset in ["train", "test"]:
            y_true = result_dict[target].get(dataset).get("y_true")
            y_pred = result_dict[target].get(dataset).get("y_pred")
            y_prob = result_dict[target].get(dataset).get("y_prob")
            model_scores = calculate_classication_scores(y_true, y_pred, y_prob)
            model_scores["target"] = target
            model_scores["train/test"] = dataset
            result_df = result_df.append(model_scores)
        
    result_df["model"] = model_name
    result_df["dataset"] = dataset_name
    index_columns = ["model", "dataset", "target", "train/test"]
    result_df = result_df[index_columns + [i for i in result_df if i not in index_columns]]
        
    return result_df
```

## Functions for SHAP values

### Obtain SHAP values from Tree Explainer, and then calculate contributions

```python
def inv_logit(x):
    """Function that converts SHAP values to probabilities"""
    return 1/(1+np.exp(-x))
```

```python
def calculate_SHAP_and_probability(model, test_set, target_vars, this_target):
    
    # Predictions
    r1 = pd.DataFrame(model.predict(test_set.drop(target_vars, axis = 1)), columns = ["prediction"])
    r2 = pd.DataFrame(model.predict_proba(test_set.drop(target_vars, axis = 1)))[[1]].rename(columns = {1: "out_prob"})
    prediction = pd.concat([r1, r2], axis = 1)
    
    # SHAP values
    shap_Explainer = shap.TreeExplainer(model)
    
    shap_values_raw = shap_Explainer.shap_values(test_set.drop(target_vars, axis = 1))
    
    feature_names = [i for i in test_set.drop(target_vars, axis = 1).columns]
    shap_values = pd.DataFrame(shap_values_raw, columns = feature_names)
    shap_values["base_value"] = shap_Explainer.expected_value
    shap_values["out_value"] = shap_values.sum(axis = 1)
    shap_values = shap_values[["out_value", "base_value"] + feature_names]
    
    output_df = pd.concat([prediction, shap_values], axis = 1)
    
    return output_df, shap_Explainer
```

```python
def from_shap_to_probablity(df):
    output_df = df.copy()
    
    ## SHAP values to probabilities START
    feature_columns = [i for i in output_df.columns if i not in ["prediction", "out_prob", "out_value", "base_value"]]
    output_df["base_prob"] = inv_logit(output_df["base_value"])
    
    output_df["distance_SHAP"] = output_df["out_value"] - output_df["base_value"]
    output_df["distance_probability"] = output_df["out_prob"] - output_df["base_prob"]    
    output_df["distance_coefficient"] = output_df["distance_SHAP"]/output_df["distance_probability"]
    
    for i in feature_columns:
        #new_column_name = i + "_contrib"
        #output_df[i + "_contrib"] = output_df[i]/output_df["distance_coefficient"]
        output_df[i] = output_df[i]/output_df["distance_coefficient"]
        
    output_df = output_df.drop(["prediction", "out_value", "base_value", "distance_SHAP", "distance_probability", "distance_coefficient"], axis = 1)
        
    output_df["out_prob_calculated"] = output_df.drop(["out_prob"], axis = 1).sum(axis = 1)
    output_df = output_df[["out_prob", "out_prob_calculated", "base_prob"] + feature_columns]
    return output_df
```

```python
def client_resampling_2019(df, distribution = {"No IBP": 0.60, "IBP": 0.40}, sampling_mode = "Oversampling"):

    if sampling_mode == "Oversampling":
    
        weight_IBP = distribution["IBP"]
        weight_No_IBP = distribution["No IBP"]
        
        client_count = testset_to_use.groupby(["month"]).agg(count_all = ("client_type","count"), count_IBP = ("client_type","sum")).reset_index()
        client_count["count_No_IBP"] = client_count["count_all"] - client_count["count_IBP"]
        client_count["count_No_IBP_expected"] = client_count["count_IBP"]*weight_No_IBP/weight_IBP
        client_count["count_No_IBP_expected"] = client_count["count_No_IBP_expected"].astype(int)
        client_count["count_No_IBP_to_add"] = client_count["count_No_IBP_expected"] - client_count["count_No_IBP"]
        client_count.loc[client_count["month"] == 12, "count_No_IBP_expected"] = client_count.loc[client_count["month"] == 12, "count_No_IBP"]
        client_count.loc[client_count["month"] == 12, "count_No_IBP_to_add"] = 0
        client_count
                
        resampling_df = pd.DataFrame()
        ## Month
        for i in np.arange(1, 13): 
            n_expected = client_count.loc[client_count["month"] == i, "count_No_IBP_expected"].values[0]
            n_to_add = client_count.loc[client_count["month"] == i, "count_No_IBP_to_add"].values[0]
            print("Month {} - Expected: {} Add: {}".format(i, n_expected, n_to_add))
            resampled = pd.concat([df.loc[(df["month"] == i)],
                                   df.loc[(df["month"] == i) & (df["client_type"] == 0)].sample(n_to_add, replace = True)],
                     axis = 0)
            resampling_df = resampling_df.append(resampled)
            
        return resampling_df
```

```python
def variable_summary(df):
    
    ### Touchpoints
    touchpoint_top3 = pd.DataFrame((df[touchpoints_to_select] >= 8).sum()/df[touchpoints_to_select].count()).rename(columns = {0: "top3"})
    touchpoint_describe = df[touchpoints_to_select].describe().transpose()
    touchpoint_describe["not_null_pct"] = touchpoint_describe["count"]/df.shape[0]
    touchpoint_summary = pd.concat([touchpoint_describe, touchpoint_top3], axis = 1)
    touchpoint_summary["group"] = "Touchpoint"
    
    #### Issues
    #issues_summary = pd.DataFrame(df[issue_variables].sum()/df.shape[0]).rename(columns = {0: "not_null_pct"})
    #issues_summary["group"] = "Issues"
    #issues_summary
    
    ### Aggregation
    aggr_prefix_list = ["cabin", "haul", "operating_airline_code", "gender_survey", "group_age", "tier_level", "cla_400_lounge_t_f"
     , "tvl_journey_reason_aggr", "departure_time_range", "fleet_aggr", "country_aggr", "booking_channel_aggr"]#, "used"]
    
    aggr_vars_list = [] 
    
    aggr_vars_dict = {}
    
    for prefix in aggr_prefix_list:
        variables = [i for i in df.columns if prefix in i and i not in touchpoints_to_select + target_vars + issue_variables]
        [aggr_vars_list.append(i) for i in variables]
        #print(prefix)
        #print(variables)
        aggr_pct = pd.DataFrame(df[variables].sum()/df.shape[0]).rename(columns = {0: "summary_dict"}).round(2).to_dict()#.get(prefix)
        aggr_vars_dict[prefix] = aggr_pct
    aggr_summary = pd.DataFrame(aggr_vars_dict).transpose()
    aggr_summary["group"] = "aggregation"
    
    ### Other
    other_summary = df[[i for i in df if i not in touchpoints_to_select + target_vars + issue_variables +  aggr_vars_list]].describe().transpose()
    other_summary["not_null_pct"] = other_summary["count"]/df.shape[0]
    other_summary["group"] = "Other"
    other_summary
    
    overall_summary = pd.concat([touchpoint_summary, aggr_summary, other_summary], axis = 0) #issues_summary
    overall_summary = overall_summary[["group"] + [i for i in overall_summary if i != "group"]]
    return overall_summary
```
